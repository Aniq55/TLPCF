```python
models = ["JODIE", "TGAT", "TGN", "CAWN"]
datasets = ["uci", "wikipedia", "reddit"]
metrics = ["AU-ROC", "AP"]
colors = ["#264653", "#2a9d8f", "#e9c46a", "#f4a261"]
distortions = ['intense', 'shuffle']
modes= ['transductive', 'inductive']


distortions_ = {
    "transductive": {"intense": "Intense", "shuffle": "Shuffle"},
    "inductive": {"intense": "Intense_inductive", "shuffle": "Shuffle_inductive"}
}



for mode in modes:
    for distortion in distortions:
        for metric in metrics:

            metric_means = { }
            for model in models:
                metric_means[model] = [data[model][dataset][distortions_[mode][distortion]][metric]["mean"] - data[model][dataset][mode][metric]["mean"] for dataset in datasets]

            metric_ci = {}
            for model in models:
                metric_ci[model] = [-data[model][dataset][distortions_[mode][distortion]][metric]["ci"] - data[model][dataset][mode][metric]["ci"] for dataset in datasets]

            x = np.arange(len(datasets))  # the label locations
            width = 0.2  # the width of the bars
            multiplier = 0

            fig, ax = plt.subplots(layout='constrained', figsize=(5,2.5))

            idx = 0
            for attribute, measurement in metric_means.items():
                offset = width * multiplier
                rects = ax.bar(x + offset, measurement, width, label=attribute, color=colors[idx%4])
                # ax.bar_label(rects, padding=3)
                multiplier += 1
                idx+=1

            multiplier = 0
            for attribute, measurement in metric_ci.items():
                offset = width * multiplier
                rects = ax.scatter(x + offset, measurement, color='k', s=10, marker='d' )
                multiplier += 1


            ax.set_xticks(x + width, datasets, fontsize=16)
            ax.legend(loc='best', ncols=2, fontsize=12)
            ax.tick_params(axis='y', which='major', labelsize=13)
            
            plt.savefig(f"/home/chri6578/Documents/gttp/barplots/{mode}_{distortion}_{metric}.pdf", bbox_inches="tight", dpi=300)
```
